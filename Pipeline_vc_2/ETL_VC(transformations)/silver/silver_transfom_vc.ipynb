```python
# silver_transform_vc.py
from pyspark.sql.functions import col, trim, upper


# Load the bronze table as a DataFrame
bronze_investments_vc = spark.table("workspace_elena_vc.default.bronze_investments_vc")

"""
transform Bronze DataFrame Ð² Silver:
    - deduplicates
    - clean
    - uppercase
"""
silver_investments_vc = (
        bronze_investments_vc.dropDuplicates()
                 .withColumn("investment_name", trim(upper(col("investment_name"))))
                 .withColumn("investor_name", trim(upper(col("investor_name"))))
    )

silver_investments_vc.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace_elena_vc.default.silver_investments_vc")
   
```
