```python
# bronze_load_vc.py
from pyspark.sql.functions import current_timestamp
"""
load from ingested table Ð² Bronze DataFrame.
"""
df_bronze = spark.table("workspace_elena_vc.default.investments_vs_froms3_ingested")

# Clean column names: remove leading/trailing spaces and replace spaces with underscores
df_bronze_clean = df_bronze.toDF(
    *[col.strip().replace(" ", "_") for col in df_bronze.columns]
)

# add column for audit
bronze_investments_vc = df_bronze_clean.withColumn("ingested_at", current_timestamp())

bronze_investments_vc.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace_elena_vc.default.bronze_investments_vc")
```
